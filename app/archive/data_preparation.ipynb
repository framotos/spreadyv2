```python
import yfinance as yf
import pandas as pd

# List of S&P 500 company tickers
sp500_tickers = sp_500["Symbol"][:500]
# Initialize an empty DataFrame to store financial data
financial_data = {}

# Loop through each ticker and retrieve financial data
for ticker in sp500_tickers:
    company = yf.Ticker(ticker)
    # Retrieve balance sheet, income statement, and cash flow
    balance_sheet = company.balance_sheet
    income_statement = company.financials
    cash_flow = company.cashflow
    financial_data[ticker] = {"balance_sheet": balance_sheet,
                              "income_statement": income_statement,
                              "cash_flow": cash_flow
    }
```

```python
from collections import defaultdict

# Dictionary to count occurrences of each KPI across companies
kpi_occurrences = defaultdict(int)

# Count each KPI's appearance across all companies' balance sheets in the 'financial_data'
for company_data in financial_data.values():
    balance_sheet = company_data.get("balance_sheet")
    if balance_sheet is not None:
        for kpi in balance_sheet.index:
            kpi_occurrences[kpi] += 1

# Convert the KPI occurrences into a DataFrame for better readability
kpi_occurrences_df = pd.DataFrame(list(kpi_occurrences.items()), columns=["KPI", "Count"]).sort_values(by="Count", ascending=False)
```

```python
data_list = []

for ticker, statements in financial_data.items():
    for statement_type, statement_df in statements.items():
        # Transpose the DataFrame to make dates rows
        statement_df = statement_df.transpose().reset_index()
        
        # Rename columns for clarity
        statement_df.rename(columns={'index': 'Date'}, inplace=True)
        
        # Add columns for Ticker (Company) and Statement Type
        statement_df['Company'] = ticker
        statement_df['Statement Type'] = statement_type.replace('_', ' ').title()  # e.g., "balance_sheet" -> "Balance Sheet"
        
        # Melt the DataFrame to get 'Metric' and 'Value' columns
        melted_df = statement_df.melt(id_vars=['Company', 'Date', 'Statement Type'], 
                                      var_name='Metric', 
                                      value_name='Value')
        
        # Append the melted DataFrame to the list
        data_list.append(melted_df)

# Concatenate all the melted DataFrames in the list into a single DataFrame
simplified_financial_data_df = pd.concat(data_list, ignore_index=True)
simplified_financial_data_df.columns = [col.lower().replace(' ', '_') for col in simplified_financial_data_df.columns]

# Display the resulting DataFrame
simplified_financial_data_df.head()
```

```python
simplified_financial_data_df['date'] = pd.to_datetime(simplified_financial_data_df['date'])
simplified_financial_data_df['value'] = pd.to_numeric(simplified_financial_data_df['value'], errors='coerce')  # This will convert 'NaN' strings to actual NaNs
```
