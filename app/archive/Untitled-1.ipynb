```python
import time
import requests
import pandas as pd

# Load the S&P 500 data
sp_500 = pd.read_csv("constituents.csv")

# Dictionary to store the fetched data
all_data = {}

# Your Alpha Vantage API key
api_key = "PTZKAD1OZ0SMA3BX"

# Loop through each ticker symbol
for ticker in sp_500["Symbol"]:
    url = f'https://www.alphavantage.co/query?function=INCOME_STATEMENT&symbol={ticker}&apikey={api_key}'
    try:
        r = requests.get(url)
        data = r.json()
        # Store the data in the dictionary
        all_data[ticker] = data
    except Exception as e:
        print(f"An error occurred for {ticker}: {e}")
        time.sleep(12)
```

```python
filtered_data = {key: value for key, value in all_data.items() if "Information" not in value}
```

```python
# Save the data to a JSON file
import json
with open("sp500_income_statements.json", "w") as f:
    json.dump(filtered_data, f)

print("Data fetching completed!")
```

```python
# Dictionary to count occurrences of each key
key_occurrences = {}

# Loop through each company and their annual reports
for company, details in filtered_data.items():
    if "annualReports" in details:
        for report in details["annualReports"]:
            for key in report.keys():
                key_occurrences[key] = key_occurrences.get(key, 0) + 1

import pandas as pd

# Convert to a DataFrame for a clean display
key_occurrences_df = pd.DataFrame.from_dict(
    key_occurrences, orient="index", columns=["Occurrences"]
).sort_values(by="Occurrences", ascending=False)
```
